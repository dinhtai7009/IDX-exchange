# Load in CSV's, Remove Outliers, and Select Features

```python
cols = ["LivingArea", "YearBuilt", "ParkingTotal", "BathroomsTotalInteger", "GarageSpaces", "AttachedGarageYN", "PoolPrivateYN", "ViewYN", "BedroomsTotal", "Stories", "AssociationFee"
, "AssociationFeeFrequency_Monthly", "AssociationFeeFrequency_None", "AssociationFeeFrequency_Quarterly", "AssociationFeeFrequency_SemiAnnually"
, 'CountyOrParish_Amador', 'CountyOrParish_Butte', 'CountyOrParish_Calaveras', 'CountyOrParish_Colusa', 'CountyOrParish_Contra Costa', 'CountyOrParish_El Dorado', 'CountyOrParish_Foreign Country', 'CountyOrParish_Fresno', 'CountyOrParish_Glenn', 'CountyOrParish_Humboldt', 'CountyOrParish_Imperial', 'CountyOrParish_Inyo', 'CountyOrParish_Kern', 'CountyOrParish_Kings', 'CountyOrParish_Lake', 'CountyOrParish_Lassen', 'CountyOrParish_Los Angeles', 'CountyOrParish_Madera', 'CountyOrParish_Marin', 'CountyOrParish_Mariposa', 'CountyOrParish_Mendocino', 'CountyOrParish_Merced', 'CountyOrParish_Mono', 'CountyOrParish_Monterey', 'CountyOrParish_Napa', 'CountyOrParish_Nevada', 'CountyOrParish_Orange', 'CountyOrParish_Other County', 'CountyOrParish_Other State', 'CountyOrParish_Placer', 'CountyOrParish_Plumas', 'CountyOrParish_Riverside', 'CountyOrParish_Sacramento', 'CountyOrParish_San Benito', 'CountyOrParish_San Bernardino', 'CountyOrParish_San Diego', 'CountyOrParish_San Francisco', 'CountyOrParish_San Joaquin', 'CountyOrParish_San Luis Obispo', 'CountyOrParish_San Mateo', 'CountyOrParish_Santa Barbara', 'CountyOrParish_Santa Clara', 'CountyOrParish_Santa Cruz', 'CountyOrParish_Shasta', 'CountyOrParish_Siskiyou', 'CountyOrParish_Solano', 'CountyOrParish_Sonoma', 'CountyOrParish_Stanislaus', 'CountyOrParish_Sutter', 'CountyOrParish_Tehama', 'CountyOrParish_Trinity', 'CountyOrParish_Tulare', 'CountyOrParish_Tuolumne', 'CountyOrParish_Ventura', 'CountyOrParish_Yolo', 'CountyOrParish_Yuba'
, 'StateOrProvince_CA', 'StateOrProvince_NV', 'StateOrProvince_OS', 'StateOrProvince_TN']
X_train = train_df[cols]
Y_train = train_df["ClosePrice"]
X_test = test_df[cols]
Y_test = test_df["ClosePrice"]
```

# Decision Tree Regressor

```python
#initialize decision tree regressor
tree = DecisionTreeRegressor(random_state=30)
```

```python
tree.fit(X_train, Y_train)
preds = tree.predict(X_test)
print(r2_score(Y_test, preds))
print(root_mean_squared_error(Y_test, preds))
```

```python
#decision tree with complexity alpha parameter
second_tree = DecisionTreeRegressor(random_state=30, ccp_alpha = 0.0001)
second_preds = second_tree.fit(X_train, Y_train).predict(X_test)
print(r2_score(Y_test, second_preds))
print(root_mean_squared_error(Y_test, second_preds))
```

Our $R^2$ is weaker than the original baseline model. I may be doing something wrong here.

# Random Forest Regressor

```python
#define random forest regressor
rf = RandomForestRegressor(random_state=30)
rf.fit(X_train, Y_train)
rf_preds = rf.predict(X_test)
print(r2_score(Y_test, rf_preds))
print(root_mean_squared_error(Y_test, second_preds))
```

```python
#decision tree with complexity alpha parameter
rf_2 = RandomForestRegressor(random_state=30, ccp_alpha = 0.1)
rf_2.fit(X_train, Y_train)
rf_2_preds = rf_2.predict(X_test)
print(r2_score(Y_test, rf_2_preds))
print(root_mean_squared_error(Y_test, rf_2_preds))
```

Random Forest massively improves $r^2$ compared to Decision Tree model. However, it only improve $r^2$ from the baseline a little bit. However, it is computationally more expensive to do than decision trees and the linear baseline model. It takes the longest to execute these cells.
